## 추천 페이지의 페이지 num 마다의 정보 가져오기

```python
# 브라우저 꺼짐 방지
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

# 불필요한 에러 메세지 없애기
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# Selenium 웹 드라이버 서비스 설정, 실행
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 페이지를 순회하면서 데이터 수집
for page_num in range(638, 1000):  # 1부터 250페이지까지
    url = f'https://www.10000recipe.com/recipe/list.html?order=reco&page={page_num}'
    driver.get(url)
    time.sleep(3)  # 페이지가 완전히 로드될 때까지 2초 대기

    # 페이지 소스 가져오기
    html = driver.page_source

    # Beautifulsoup을 사용하여 파싱
    soup = BeautifulSoup(html, 'html.parser')

    # box_contents 찾기
    box_contents = soup.find_all('li', class_='common_sp_list_li')

    # atag_contents 찾기
    atag_contents = soup.find_all('a', class_='common_sp_link')
    
    # 현재 페이지의 box_contents를 해당 페이지 번호에 맞는 변수에 저장
    exec(f'box_contents{page_num} = box_contents')

    # 현재 페이지의 atag_contents를 해당 페이지 번호에 맞는 변수에 저장
    exec(f'atag_contents{page_num} = atag_contents')

    # 페이지당 데이터를 파일로 저장하거나 원하는 작업을 수행할 수 있습니다.
    # 이 코드에서는 간단히 출력합니다.
#     print(f"Page {page_num}: Found {len(box_contents{page_num})} box_contents and {len(atag_contents{page_num})} atag_contents")
    print(f"Page {page_num}: Found {len(eval('box_contents' + str(page_num)))} box_contents and {len(eval('atag_contents' + str(page_num)))} atag_contents")
    
# 작업이 완료되면 웹 드라이버를 종료합니다.
driver.quit()
```

## url로 접근하여 데이터 프레임 만들기

```python
# 브라우저 꺼짐 방지
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

# 불필요한 에러 메세지 없애기
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 알림 대응 설정 추가
chrome_options.add_argument("--disable-notifications")

# 알림 창 자동 수락 설정 추가
chrome_options.add_argument("--disable-popup-blocking")
chrome_options.add_argument("--enable-automation")

# UnexpectedAlertPresentException 발생 시 알림 처리
chrome_options.add_argument("--disable-infobars")
chrome_options.add_argument("--disable-web-security")
chrome_options.add_argument("--allow-running-insecure-content")
chrome_options.add_argument("--ignore-certificate-errors")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("--disable-save-password-bubble")
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("--disable-default-apps")
chrome_options.add_argument("--disable-extensions-file-access-check")
chrome_options.add_argument("--disable-search-geolocation-disclosure")

# 알림 창 자동 수락 설정 추가
chrome_options.add_experimental_option("prefs", {"profile.default_content_setting_values.notifications": 1})

# Selenium 웹 드라이버 서비스 설정, 실행
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 기존의 implicitly_wait 제거 또는 한 번만 설정
driver.implicitly_wait(5)

# 모든 레시피 재료 정보를 저장할 큰 리스트
review_list = []
time_list = []
portion_list = []
all_recipes_ingredients = []
p_num = 0

# 제거할 인덱스를 저장할 리스트
indexes_to_remove = []

for index, url in enumerate(url_list):
    driver.get(url)
    time.sleep(4)
#     driver.execute_script("window.scrollTo(0, 1200);")
    p_num += 1
    
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '#divConfirmedMaterialArea'))
        )
        
        # 현재 레시피의 모든 재료를 저장할 리스트
        current_recipe_ingredients = []
        
        divs = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, 'ingre_list_name'))
        )

         # 재료명을 가져와서 리스트에 추가
        for div in divs:
            try:
                ingre_list_name = WebDriverWait(div, 10).until(
                    EC.presence_of_element_located((By.TAG_NAME, 'a'))
                ).text

                current_recipe_ingredients.append(ingre_list_name)
            
            except NoSuchElementException:
                print("태그 'a'를 찾을 수 없습니다.")
                
        # 현재 레시피의 조리 시간 및 인분 가져오기
        summary_info = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'view2_summary_info'))
        )
        portion = summary_info.find_element(By.CLASS_NAME, 'view2_summary_info1').text.strip() if summary_info.find_element(By.CLASS_NAME, 'view2_summary_info1') else None
        t_time = summary_info.find_element(By.CLASS_NAME, 'view2_summary_info2').text.strip() if summary_info.find_element(By.CLASS_NAME, 'view2_summary_info2') else None
        
        # 현재 레시피의 리뷰 수 가져오기
        review_spans = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.reply_tit span'))
        )
        
        review_count = 0
        for span in review_spans:
            r_text = span.text.strip()  # 텍스트 앞뒤 공백 제거
            if r_text.isdigit():  # 텍스트가 숫자인지 확인
                review_count += int(r_text)  # 숫자라면 정수로 변환하여 누적
    
        # 데이터를 리스트에 추가
        time_list.append(t_time)
        portion_list.append(portion)
        review_list.append(review_count)
        
        # 현재 레시피의 재료들을 하나의 문자열로 결합하여 리스트에 추가
        all_recipes_ingredients.append(', '.join(current_recipe_ingredients))
        
    except TimeoutException:
        print("제한 시간 초과")
        # 제한 시간이 초과되면 해당 URL을 url_list에서 제거
        indexes_to_remove.append(index)
    except NoSuchElementException:
        print("요소를 찾을 수 없음")
        indexes_to_remove.append(index)
    except UnexpectedAlertPresentException:
        
        print("알 수 없는 경고가 발생")
        # 알 수 없는 경고가 발생할 경우 해당 URL을 url_list에서 제거
        indexes_to_remove.append(index)
        continue
        
    print(f"page is OK{p_num}")
# 작업이 완료되면 웹 드라이버를 종료합니다.
driver.quit()
# url_list에서 해당하는 인덱스를 제거
for index in sorted(indexes_to_remove, reverse=True):
    del url_list[index]
    
# img_list에서 해당하는 인덱스를 제거
for index in sorted(indexes_to_remove, reverse=True):
    del img_list[index]

# views_list에서 해당하는 인덱스를 제거
for index in sorted(indexes_to_remove, reverse=True):
    del views_list[index]

# recipe_name_list에서 해당하는 인덱스를 제거
for index in sorted(indexes_to_remove, reverse=True):
    del recipe_name_list[index]



# s_df.to_csv('recipes_data.csv', index=False)
```

```python
import pandas as pd
import re

# 데이터프레임 생성
df = pd.DataFrame({'ingredients': ['굴소스 some text', '올리고당 some text', '고추장 some text', '간장 some text', '물 some text']})

# 특정 텍스트 제거할 패턴 설정
texts_to_remove = ['굴소스', '올리고당', '고추장', '간장', '물']

# 패턴 적용하여 특정 텍스트 삭제하는 람다 함수 정의
remove_texts = lambda row: re.sub('|'.join(texts_to_remove), '', row)

# apply 메서드를 사용하여 람다 함수 적용
df['ingredients'] = df['ingredients'].apply(remove_texts)

# 업데이트된 데이터프레임 출력
print(df)
```

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({'ingredients': ['소고기등심, 새싹채소, , , , , , , ,']})

# 쉼표를 기준으로 문자열을 분할하고 빈 문자열을 제거하여 다시 결합
df['ingredients'] = df['ingredients'].apply(lambda x: ','.join(filter(lambda x: x.strip(), x.split(','))))

# 업데이트된 데이터프레임 출력
print(df)
```
