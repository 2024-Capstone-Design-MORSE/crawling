```python
import requests
import uuid
import time
import json
api_url = 'https://4tr7b5lf7u.apigw.ntruss.com/custom/v1/29214/5297e0a771dd8193d310e5fd6b2695648dd9fdeac6e8b2c38519330cea9d78c6/general'
secret_key = 'ckNvamlrdG9WTmlsRXF3bWZRcFdlVlFPeUdpTXZaWk4='
image_file = 'test_receipt.jpg'
request_json = {
    'images': [
        {
            'format': 'jpg',
            'name': 'test_receipt'
        }
    ],
    'requestId': str(uuid.uuid4()),
    'version': 'V2',
    'timestamp': int(round(time.time() * 1000))
}
payload = {'message': json.dumps(request_json).encode('UTF-8')}
files = [
  ('file', open(image_file,'rb'))
]
headers = {
  'X-OCR-SECRET': secret_key
}
response = requests.request("POST", api_url, headers=headers, data = payload, files = files)
print(response.text.encode('utf8'))
```

```python
result = response.json()
with open('result.json', 'w', encoding='utf-8') as make_file:
    json.dump(result, make_file, indent="\t", ensure_ascii=False)
```
